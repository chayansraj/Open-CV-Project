# Blending and Pasting Images

```python
# Blending is a method of combining two or images and add some weight of each image on the final result. We use the 
# 'addWeighted' function to achieve this task. 
```


# Formula to calculate final pixel of the image

## new_pixel = (alpha * first_image_pixel) + (beta * second_image_pixel) + gamma(optional)



```python
img1 = cv2.imread('Computer-Vision-with-Python/DATA/dog_backpack.png')
```

```python
img2 = cv2.imread('Computer-Vision-with-Python/DATA/watermark_no_copy.png')
```

```python
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
img1.shape
```

```python
img2.shape
```

```python
# Resizing to blend the images of same size

img1 = cv2.resize(img1,(1200,1200))
img2 = cv2.resize(img2,(1200,1200))
```

```python
blended = cv2.addWeighted(img1, 1, img2, 0.1,0.2) # addWeighted works only when the images are of the same size
```

```python
plt.imshow(blended)
```

```python
# OVERLAY SMALL IMAGE ON TOP OF LARGER IMAGE
# NUMPY REASSIGNMENT 
```

```python
img1 = cv2.imread('Computer-Vision-with-Python/DATA/dog_backpack.png')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
```

```python
img2 = cv2.imread('Computer-Vision-with-Python/DATA/watermark_no_copy.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img2 = cv2.resize(img2, (600,600))
```

```python
img1.shape
```

```python
img2.shape
```

```python
large_img = img1
small_img = img2
```

```python
large_img.shape
```

```python
small_img.shape
```

```python
x_offset = 0
y_offset = 0
```

```python
x_end = x_offset + small_img.shape[1]  #Because numpy treats columns as the x-axis of the image
y_end = y_offset + small_img.shape[0]  #Because numpy treats rows as the y-axis of the image
```

```python
large_img[x_offset:x_end, y_offset:y_end] = small_img
```

```python
plt.imshow(large_img)
```
