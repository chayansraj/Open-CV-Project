```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img, cmap='gray'):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
reeses = cv2.imread('Computer-Vision-with-Python/DATA/reeses_puffs.png',0)
```

```python
cereals = cv2.imread('Computer-Vision-with-Python/DATA/many_cereals.jpg',0)
```

```python
sift = cv2.xfeatures2d.SIFT_create()
```

```python
kp1, des1 = sift.detectAndCompute(reeses,None)
kp2, des2 = sift.detectAndCompute(cereals,None)
```

```python
# FLANN BASED PARAMETERS 
FLANN_INDEX_KDTREE = 0
index_params = dict(algorithm = FLANN_INDEX_KDTREE, trees = 5)
search_params = dict(checks = 50)
```

```python
flann = cv2.FlannBasedMatcher(index_params, search_params)
```

```python
matches = flann.knnMatch(des1,des2, k=2)
```

```python
matches_mask = [[0,0] for i in range(len(matches))]
```

```python
matches
```

```python

# LESS DISTANCE == BETTER MATCH
# RATIO MATCH1 < 70% MATCH 2
for i, (match1,match2) in enumerate(matches):
    # IF MATCH1 DISTANCE IS LESS THAN 70% OF MATCH2 DISTANCE, THEN DESCRIPTOR WAS A GOOD MATCH
    if match1.distance < 0.7 * match2.distance:
        matches_mask[i] = [1,0]
```

```python
draw_params = dict(matchColor =(0,255,0), singlePointColor= (255,0,0), matches_mask = matches,flags=0)
```

```python
draw_params
```

```python
flann_matches = cv2.drawMatchesKnn(reeses,kp1,cereals,kp2,matches,None,**draw_params)
```

```python
display(flann_matches)
```
