```python
import cv2

cap = cv2.VideoCapture(0)

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))


# TOP LEFT CORNER OF RECTANGLE
x = width // 2
y = height // 2


# THE SIZE OF THE RECTANGLE AND MAKING THE BOTTOM RIGHT CORNER 
w = width // 4
h = height // 4


while True:
    
    ret,frame = cap.read()
    cv2.rectangle(frame, (x,y), (x+w,y+h), color=(0,0,255), thickness = 3) # lATER THESE CO-ORDINATES WILL BE DETECTED BY OBEJCT DETECTION 
    cv2.imshow('frame',frame)
    
    if cv2.waitKey(1) & 0xFF == 27:
        break
        
cap.release()
cv2.destroyAllWindows()
```
