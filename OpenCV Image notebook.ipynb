```python
#We can also open image with different color graphics 
img_gray = cv2.imread('Computer-Vision-with-Python/DATA/00-puppy.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img_gray, cmap='gray') #Because we are displaying it with matplotlib and we have to mention the colormapping
```

```python
img_gray.shape
```

```python
# Resizing the image 
resized_image = cv2.resize(img_gray, (1000,500))
```

```python
plt.imshow(resized_image)
```

```python
#To expand the canvas size of the photo
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
ax.imshow(img_gray)
```
