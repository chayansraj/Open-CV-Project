# What is a histogram? 
## Histogram is basically a visual representation of the distribution of a continuous feature.
## For image histogram, we can display the frequency of values for colors. Each of the color channels have values between 0-255. We can plot these as 3 histogram on top of each other and then we can check how much of each channel is there.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dark_horse = cv2.imread('Computer-Vision-with-Python/DATA/horse.jpg') # For original BGR OpenCV
show_horse = cv2.cvtColor(dark_horse, cv2.COLOR_BGR2RGB) # For showing through matplotlib

rainbow = cv2.imread('Computer-Vision-with-Python/DATA/rainbow.jpg')
show_rainbow = cv2.cvtColor(rainbow, cv2.COLOR_BGR2RGB)

blue_bricks = cv2.imread('Computer-Vision-with-Python/DATA/bricks.jpg')
show_bricks = cv2.cvtColor(blue_bricks, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(show_horse)
```

```python
hist_bricks = cv2.calcHist([blue_bricks], channels=[0], mask=None, histSize=[256], ranges=[0,256])
```

```python
plt.plot(hist_bricks)
```

```python
hist_horse = cv2.calcHist([dark_horse], channels=[0], mask=None, histSize=[256], ranges=[0,256])
```

```python
plt.plot(hist_horse)
```

```python
img = blue_bricks
```

```python
color = ('b','g','r')

for i,col in enumerate(color):
    histr = cv2.calcHist([img], [i], None, [256], [0,256])
    plt.plot(histr, color=col)
    plt.xlim([0,256])
    
plt.title('Histogram for blue bricks')
```

```python
color = ('b','g','r')

for i,col in enumerate(color):
    histr = cv2.calcHist([dark_horse], [i], None, [256], [0,256])
    plt.plot(histr, color=col)
    plt.xlim([0,50])
    plt.ylim([0,500000])
    
plt.title('Histogram for Dark horse')
```

```python
color = ('b','g','r')

for i,col in enumerate(color):
    histr = cv2.calcHist([rainbow], [i], None, [256], [0,256])
    plt.plot(histr, color=col)
    plt.xlim([0,256])
    
plt.title('Histogram for Rainbow')
```

# Histogram Equalization
## We use the above method to tweak the contrast of an image just like we use gamma correction for adjusting brightness. Applying HE will reduce the color depth of shades of gray areas.

```python
gorilla = cv2.imread('Computer-Vision-with-Python/DATA/gorilla.jpg',0)
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
display_img(gorilla)
```

```python
gorilla.shape
```

```python
hist_gorilla = cv2.calcHist([gorilla],channels=[0],mask=None, histSize=[256], ranges=[0,256])
```

```python
plt.plot(hist_gorilla)
```

```python
eq_hist_gorilla = cv2.equalizeHist(gorilla)
```

```python
display_img(eq_hist_gorilla)
```

```python
hist_eq = cv2.calcHist([eq_hist_gorilla],channels=[0],mask=None, histSize=[256], ranges=[0,256])
```

```python
plt.plot(hist_eq)
```
